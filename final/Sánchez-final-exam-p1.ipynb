$$V(r) = 4\epsilon \left[ \left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6 \right]$$

### Solution

(a) Create a python function to compute this potential $V(r)$ as a function of $r$ for fiducial values of $\epsilon$ and $\sigma$.

Then, create an appropriate vector for $r$ and plot this potential $V(r)$ versus $r$.


```python
# Defining the function
def v_potential(r, epsilon, sigma):
    """
    This function calculates the Lennard-Jones potential energy between two particles 
    based on their separation distance r, the depth of the "potential well" epsilon, 
    and the characteristic distance sigma where the potential reaches its minimum.
    Input:
        r (float) -> Separation distance between two particles.
        epsilon (float) -> Depth of the potential well (energy scale).
        sigma (float) -> Characteristic distance where the potential is minimum.

    Output:
        v (float) -> Lennard-Jones potential.

    Author: R.S.S.G
    Date created: 7/12/2024
    """
    rho = (sigma/r)
    p = rho**12 - rho**6
    v = 4*epsilon*p
    return v
```

```python
# Defining the values for epsilon and sigma, I will use one for both as it will give us characteristic constants
# However, if changed for realistic values this could represent visualization problems as real values are too small (x10^-10)
# The whole notebook would not have any problem if these values are changed, just comment or adjust the limits in the plots.

epsilon = 1
sigma = 1

# Generating a vector with the distances
r_values = np.linspace(1e-3, 3*sigma, 1000)

# Evaluate our potential
v_values = v_potential(r_values, epsilon, sigma)

# We know the minimun from other physics courses, and we're going to derive it later in this notebook
# Let's use it to indicate the minimum
r_v_min = (2**(1/6))*sigma
v_min = v_potential(r_v_min, epsilon, sigma) # or we could just use epsilon here
```

```python
plt.figure(figsize=(10,6))

plt.plot(r_values, v_values, color = "k", label = "Lennard-Jones potential")
plt.plot(r_v_min, v_min, color = "r", label = r"Minimum $(2^{\frac{1}{6}}\, \sigma, - \epsilon ) $", marker = "o")

plt.title(r"$V(r) \, vs \, r$")
plt.xlabel(r"$r$ [Length]")
plt.ylabel(r"$V(r)$ [Energy]")
plt.legend(loc="upper right", ncol = 1)
plt.xlim(0,np.max(r_values))
plt.ylim(-1.15*(epsilon),1.5*epsilon)

plt.axvline(r_v_min, linestyle = "--", color = "r", alpha = 0.4)
plt.axhline(v_min, linestyle = "--", color = "r", alpha = 0.4)

plt.show()
plt.close()
```

(b) Use sympy to compute the force: $F(r)=-\frac{dV}{dr}$ experienced by a particle at a distance $r$.

Then, lambdify your expression and plot the $F(r)$ versus $r$.

```python
# Define symbols 
r = sp.symbols("r")
v = v_potential(r, epsilon, sigma)

# Get F
f = - sp.diff(v,r)
print("The force is given by")
display(f)
# Lambidify the function
f_function = sp.lambdify(r,f, "numpy")

# Getting numeric values to plot
f_values = f_function(r_values) 
```

```python
plt.figure(figsize=(10,6))

plt.plot(r_values, f_values, color = "b", label = "Lennard-Jones force")

plt.title(r"$F (r ) \, vs \, r$")
plt.xlabel(r"$r$ [Length]")
plt.ylabel(r"$F(r)$ [Force]")
plt.legend(loc="upper right", ncol = 1)
plt.xlim(0,np.max(r_values))
plt.ylim(1.15*(np.min(f_values)),2)
plt.show()
plt.close()
```

(c) Find the maximum possible value of the attractive force, $F_{max}$, and the distance, $r_{max}$, at which the particle would be subjected to this force.

**Answer:** <br>
If we want the maximum value of the magnitud of the attractive force, this corresponds to the minimum of the total force, and to get it we can make the first derivative of $F(r)$ equal to zero and solve for r. <br>
$$\dfrac{dF(r)}{dr} = 0$$
Notice that we should only care about $r > 0$ (and $r \in \mathbb{R}$) results as we have a central force!

```python
# Get the maximum by getting the first derivative
eq = sp.Eq(sp.diff(f,r),0)

print("The equation for the minimum would be")
display(eq)

print("The solution rmax, considering r>0,  would be")
sln = sp.solve(eq, r)
# After inspecting sln, we can take the only reasonable solution for our problem
r_f_max = sln[1]
display(r_f_max)
print("Numerically, rmax = ", float(r_f_max))


# Now, evaluating it to get the force
a_f_max = f_function(r_f_max)
print("The maximum magnitud for the attractive force would be: ", "|Fmax| = ", np.abs(float(a_f_max)))
```

Then, the distance at which the particle experiences the maximum attractive force is $$r_{max} = (\frac{26}{7})^{1/6}\, \sigma = 1.244455060259808 \, \sigma $$

**Answer:** <br>
We know, by the form of the force, that at long distances (*negative force values*) is attractive, while for short distances (*positive force values*) is repulsive. In fact, this transition from attractive to repulsive implies that the force must be zero at some equilibrium distance $r_0$ that, as we will see, correponds to the distance that minimizes the potential $V(r)$. 

```python
# As we stated, we have two ways of getting this equilibrium distance r0
# First, let's get it with the force

eq = sp.Eq(f,0)

print("The equation for zero force is")
display(eq)

print("The solution r0, considering r>0,  would be")
sln = sp.solve(eq, r)
# After inspecting sln, we can take the only reasonable solution for our problem
r_0 = sln[1]
display(r_0)
print("Numerically, r0 = ", float(r_0))
```

```python
# As we stated, we have two ways of getting this equilibrium distance r0
# Alternatively, let's get it with the potential

eq = sp.Eq(sp.diff(v),0)

print("The equation for minimum in the potential is")
display(eq)

print("The solution r0, considering r>0,  would be")
sln = sp.solve(eq, r)
# After inspecting sln, we can take the only reasonable solution for our problem
r_0 = sln[1]
display(r_0)
print("Numerically, r0 = ", float(r_0))
```

Both "methods" are equivalent as they represent the exact same thing.

Then, the equilibrium distance is $$r_0 = (2^{1/6})\, \sigma$$
Where sigma is replaced by the value given at the beggining, in this notebook we used $\sigma = 1$.

(e) Discuss the physical significance of the Lennard-Jones potential in real-world systems (e.g. the meaning of $\sigma$ and $\epsilon$).

Provide 2 examples of systems where this potential is applicable.

**Answer:**<br>
From other physics courses, we know that this potential describes the interaction between a **pair** of atoms or molecules only as a function of the distance between them. The attractive part of the potential (the term with the minus sign) comes from van der Waals interactions mainly that are present in intermediate not so large distances, while the repulsive part (the positive term) can be interpreted as a consequence of the Pauli exclusion principle that prevents fermions from occupying the same states (for example, two electron clouds getting closer, and therefore, getting repeled by each other increasingly as the distance decreases). <br>
Now, $\epsilon$ represents the depth or strength of the attractive potential well, describing how strongly the two particles attract each other, specially when separated by a distance near to $\sigma$. On the other hand, $\sigma$ represents a characteristic distance, as we can see from the equations, when $r = \sigma$ the potential $V(r)=0$. And also, we can see its appearance in the expression for $r_0$ presented previously. In some context, like molecular systems or gases, can be interpreted as the diameter or size of the particles (e.g, for argon it can represent the "effective" diameter).

For the two examples:
1) As previously mentioned, we can use this potential to model Noble gases like argon or neon as they interact very weakly through these van der Waals forces.

The **Verhulst-Pearl model** is used to study animal population dynamics. This model is also called **logistic growth model** and assumes that $P(t)$ satisfies the following ODE:


$$\frac{dP}{dt}=\alpha\,P\left(1-\frac{P}{\beta}\right)$$


where $\alpha$ is the growth rate and $\beta$ is the limiting size for the population (also called the environment carrying capacity).

### Solution

(a) Use sympy to find the general analytical solution to this ODE.

```python
# Define the symbols
p = sp.symbols("P", cls = sp.Function)
t, a, b = sp.symbols("t, a, b") 
```

```python
# Writing the ODE
ode_vp = sp.Eq(p(t).diff(t), a * p(t) * (1 - (p(t)/b)) )
# We could display the ODE
display(ode_vp) 

# Solving it
sln_vp = sp.dsolve(ode_vp)
print("The solution is given by")
display(sln_vp)
```

(b) Imagine we use this equation to study the population dynamics of a bird species re-introduced into nature, knowing that the test environment can only host $20$ individuals. We liberate $10$ individuals of this species in the wild, and we know that the growth rate is $0.2$ individuals per month. Rewrite and recompute both the ODE and its solution for this set of conditions.

```python
# Let's define our parameters
a = 0.2 # growth rate
b = 20 # limiting size of the population

# Rewriting the ODE
ode_vp_val = sp.Eq(p(t).diff(t), a * p(t) * (1 - (p(t)/b)) )

# Solving it using the initial condition P(0)
sln_vp_val = sp.dsolve(ode_vp_val, ics = {p(0):10})
print("The new solution, considering the conditions, is given by")
display(sln_vp_val)
```

(c) Lambify your expression, make a plot of the analytical solution and label it appropriately.

```python
# Lambidify the function 
p_function = sp.lambdify(t, sln_vp_val.rhs ,"numpy")

# Defining a time axis to evaluate P(t)
t_val = np.arange(0,72,0.001)

# Evaluating the function
p_val = p_function(t_val) 
```

```python
plt.figure(figsize=(10,6))

plt.plot(t_val, p_val, color = "g", label = "$P(t)$ with $\\alpha = 0.2$, \n $\\beta = 20$ & $P(0)=10$")

plt.title(r"Verhulst-Pearl model")
plt.xlabel(r"$t \, [month]$")
plt.ylabel(r"$P(t)$")

plt.axhline(b, color = "purple", linestyle = "--", alpha = 0.3)

plt.ylim(9.5, 21)

plt.legend(loc=5)
plt.show()
plt.close()
```

(e) Why does the population growth slow down as $P(t)$ approaches $\beta$ ?

How might the model change if external factors (e.g., predators, diseases, or resource scarcity) are introduced?

**Answer:** <br>
Mathematically, as $P(t)$ approaches $\beta$, the term ($1 - \frac{P(t)}{\beta}$) in the ODE vanishes, resulting in $$\dfrac{dP}{dt} \to 0$$
This factor in the ODE accounts for the environmental limitations that appear from the same growth on the population, i.e, it depends on $P(t)$. In other words, as the population increases, individuals consume more resources, which in turn slows down population growth as it approaches the limiting size $\beta$. <br>

Now, if external factors are introduced, the Verhulst-Pearl model would be likely to fail at predicting the growth of a population as this model mimics an idealized scenario. <br>
To have a better prediction, we should add some extra parameters like:
1) Death rate parameter $\eta$, this would account for the effect of the presence of predators or diseases on the population.
2) Growth and death parameters for the predators/diseases $\gamma$, $\delta$. These would account for the external factors that affect to the predators, e.g, the ammount of preys available for consumption. Or for the diseases, how easily can be transmitted or how it behaves for crowded populations, etc.
3) A time dependent limiting size factor $\beta (t)$ would account for the resource scarcity during specific situations or environmental conditions over time.

Now, we may also want to introduce stochasticity if we want to represent a more realistic scenario, as it would account for the unpredictable events that can happen in real dynamical systems. We could implement this by using Monte Carlo simulations to generate many possible outcomes.
