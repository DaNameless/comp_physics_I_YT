## Part 3. (12 points) Data analysis and visualisation (Orszag-Tang MHD vortex)

The **Orszag-Tang vortex** system describes a doubly periodic ideal gas configuration leading to 2D supersonic magnetohydrodynamical (MHD) turbulence. Although an analytical solution is not known, its simple and reproducible set of initial conditions has made it very appealing for the comparison of MHD numerical solvers. The computational domain is a **periodic box** with dimensions: $[0,2\pi]^2$, and the gas has an adiabatic index $\gamma=\frac{5}{3}$.

In code units, the initial conditions are given by:

$
\vec{v}_{code} = \left(-\sin y,\, \sin x\right) \,,\qquad
\vec{B}_{code} = \left(-\sin y,\, \sin 2x\right) \,,\qquad
\rho_{code} = 25/9\,,\qquad
p_{code}    = 5/3,
$

and the numerical simulation produces $61$ VTK files stored in:

- the **Orszag-Tang-MHD** folder: 

https://github.com/wbandabarragan/physics-teaching-data/blob/main/2D-data/Orszag_Tang-MHD.zip

which also contains the following descriptor files:

- a **units.out** file that contains the CGS normalisation values.
- a **vtk.out** file whose second column contains the times in code units.
- a **grid.out** file that contains information on the grid structure.

You can use VisIt to inspect the data. The written fields are: 

- density (rho)
- thermal pressure (prs)
- velocity_x (vx1)
- velocity_y (vx2)
- magnetic_field_x (Bx1)
- magnetic_field_y (Bx2)

**Reference paper:**
https://arxiv.org/pdf/1001.2832.pdf

"High-order conservative finite difference GLM-MHD schemes for cell-centered MHD", Mignone, Tzeferacos & Bodo, JCP (2010) 229, 5896.

(a) Create a python function that reads the **units.out** file, stores the normalisation values for length, velocity, and density, calculates the normalisation values for thermal pressure, magnetic field, and time, and returns them all into tuple objects.

**Note:**
As shown in class, the normalisation values for thermal pressure ($p_0=\rho_0\,v_0^2$), magnetic field ($B_0=\sqrt{4\,\pi\,\rho_0\,v_0^2}$), and time ($t_0=\frac{L_0}{v_0}$) can be derived from the length, velocity, and density values.

```python
# Define the directory
directory = "/home/rolando/Downloads/Yachay/7th Semester/computational_physics/exams/data/Orszag_Tang-MHD/"

# Define the filename
units = "units.out"
```

```python
def io_units(filename):
    """
    This functions reads an units.out file and returns the normalization values in cgs units.
    Input:
        filename (str) -> location and name of the file
    Output:
        All the following quantities as a tuple
        rho_0 (float) -> normalization value for density in [g/cm^3]
        v_0 (float) -> normalization value for velocity in [cm/s]
        L_0 (float) -> normalization value for length in [cm]
        p_0 (float) -> normalization value for thermal pressure [1 dyn/cm^2]
        B_0 (float) -> normalization value for magnetic flux density in [G] (gauss)
        t_0 (float) -> normalization value for time in [s]
    Author: R.S.S.G
    Date created: 18/10/2024     
    """
    # Read the file
    df = pd.read_csv(filename, sep = ",")
    # Getting the values
    rho_0 = np.array(df.loc[df["variable"]== "rho_0"]["normalisation"])
    v_0 = np.array(df.loc[df["variable"]== "v_0"]["normalisation"])
    L_0 = np.array(df.loc[df["variable"]== "L_0"]["normalisation"])
    # Calculate the other normalization values
    p_0 = rho_0*(v_0**2)
    B_0 = np.sqrt(4*np.pi*rho_0*(v_0**2))
    t_0 = L_0/v_0
    return rho_0, v_0, L_0, p_0, B_0, t_0
```

```python
# Calling our function to use these values in the next items.
rho_0, v_0, L_0, p_0, B_0, t_0 = io_units(directory+units)
```

(b) Create a python function that reads the **vtk.out** file, reads the second column, and returns the times in CGS units using $t_0$ from point (a).

```python
# Define the filename
vtk_out = "vtk.out"
```

```python
# Defining the function
def io_vtk_out(filename):
    """
    This functions reads a vtk.out file and returns the time data array in [s] (cgs units for time).
    Input:
        filename (str) -> location and name of the file
    Output:
        t_array (np.array (of) floats) -> 1D array of time-stamps.
    Author: R.S.S.G
    Date created: 18/10/2024
    """
    # Read the file
    df = pd.read_csv(filename, sep = "\\s+", header = None)
    # Getting the values in [code units]
    t_array = np.array(df[1])
    # Converting them into [s]
    t_array = t_0 * t_array
    return t_array
```

```python
t_array = io_vtk_out(directory + vtk_out)
```

(c) Create a python function that reads a VTK data file, normalises the data fields to CGS units using the values from points (a) and (b), and returns them jointly with the mesh and time information as tuple objects.

```python
def io_vtk(directory, filename):
    """
    This function reads a .vtk file and returns data fields in CGS units jointly 
    with a 2D mesh and time information as tuple objects.
    Input:
        directory (str) -> location of the file
        filename (str) -> name of the file
    Output:
        All the following data in a tuple.
        rho_cgs2D (np.array (of) floats) -> normalized 2D density array in [g/cm^3]
        vx1_cgs2D (np.array (of) floats) -> normalized 2D velocity in x dir array in [cm/s]
        vx2_cgs2D (np.array (of) floats) -> normalized 2D velocity in y dir array in [cm/s]
        Bx1_cgs2D (np.array (of) floats) -> normalized 2D magnetic flux density in x dir array in [G] 
        Bx2_cgs2D (np.array (of) floats) -> normalized 2D magnetic flux density in y dir array in [G]
        prs_cgs2D (np.array (of) floats) -> normalized 2D thermal pressure array in [1 dyn/cm^2]
        mesh_grid (tuple (of) np.meshgrid objects (two 2D np.arrays)) -> X and Y 2D meshgrid with data dimensions.
        t (float) -> specific time-stamp in [s]
    Author: R.S.S.G
    Date created: 18/10/2024
    """
    
    # Get the data into a mesh
    mesh = pv.read(directory + filename)
    # print(mesh)
    # print(mesh.array_names)
    # Get data arrays in code units:
    rho = pv.get_array(mesh, "rho", preference = "cell")
    vx1 = pv.get_array(mesh, "vx1", preference = "cell")
    vx2 = pv.get_array(mesh, "vx2", preference = "cell")
    Bx1 = pv.get_array(mesh, "Bx1", preference = "cell")
    Bx2 = pv.get_array(mesh, "Bx2", preference = "cell")
    prs = pv.get_array(mesh, "prs", preference = "cell")
    # Normalise the quantities to CGS units
    rho_cgs = rho*rho_0
    vx1_cgs = vx1*v_0
    vx2_cgs = vx2*v_0
    Bx1_cgs = Bx1*B_0
    Bx2_cgs = Bx2*B_0
    prs_cgs = prs*p_0
    
    # Reshaping them into their previous shapes
    rho_cgs2D = rho_cgs.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    vx1_cgs2D = vx1_cgs.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    vx2_cgs2D = vx2_cgs.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    Bx1_cgs2D = Bx1_cgs.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    Bx2_cgs2D = Bx2_cgs.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    prs_cgs2D = prs_cgs.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
    # Create coordinate vectors:
    x = np.linspace(mesh.bounds[0], mesh.bounds[1], mesh.dimensions[1] - 1)*L_0
    y = np.linspace(mesh.bounds[2], mesh.bounds[3], mesh.dimensions[0] - 1)*L_0
    # Generate Grid
    x_2d, y_2d = np.meshgrid(x, y)
    mesh_grid = (x_2d, y_2d)
    # Now, we want to get specific times from the t_array based on the name of the file
    # Therefore, we could use the function .strip to remove all characters except the numbers.
    ind_t = int(filename.strip("data.vtk"))
    # Using this index we can get the value
    t = t_array[ind_t]

    return rho_cgs2D, vx1_cgs2D, vx2_cgs2D, Bx1_cgs2D, Bx2_cgs2D, prs_cgs2D, mesh_grid, t
```

(e) Create a set of Python functions that loops over all the VTK simulation files and returns maps of the density field $\rho$, the kinetic energy density $E_k$, and also histograms of the density field in CGS units for all times, into a folder called "output_data".


```python
def maps_and_histograms(directory, format):
    """
    This function reads multiple .vtk files and make plots of the density field, kinetic energy density and histograms of 
    the density field in CGS units.
    Input:
        directory (str) -> location of the .vtk files
        format (str) -> Specific format for the image
    Output:
        density_i_map.eps (eps image(s)) -> Plots of density for each time step (each vtk file).
        kinetic_energy_i_map.eps (eps image(s)) -> Plots of kinetick energy density for each time step (each vtk file).
        density_i_histogram.eps (eps image(s)) -> Histogram of density field for each time step (each vtk file).
    Author: R.S.S.G
    Date created: 19/10/2024
    """
    
    # First let's define plotting functions!
    def plotting_scalar(field, mesh, i, t_i, name, format):
        # Plotting the scalar field
        plt.figure(figsize=(8,6))
    
        if name == "density":
            plt.title("Gas density $\\rho \\,[g/cm^3]$")
            z = plt.pcolormesh(*mesh, np.log10(field), vmin=-24, vmax=-23.2, cmap="cividis", shading='auto')
        else:
            plt.title("Kinetic energy density $E_k \\, [erg/cm^3]$")
            z = plt.pcolormesh(*mesh, np.log10(field), cmap="inferno_r", vmin=-19.3, vmax=-12.8, shading='auto')  
            
        # Adding a color bar with a label
        cbar = plt.colorbar(z)
        if name == "density":
            cbar.set_label("log$_{10}$($\\rho$)")
        else:
            cbar.set_label("log$_{10} (E_{k})$")
        
        plt.xlabel("$x \\, [cm]$")
        plt.ylabel("$y \\, [cm]$")
        
        # Adding a box with the time-stamp
        plt.text(0.05, 0.95, f"Time: {t_i:.3e} [s]", transform=plt.gca().transAxes,
                 fontsize=10, verticalalignment='top',
                 bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        plt.savefig(f"output_data/{name}_{i:03d}_map.{format}")
        plt.close()
        return
    
    def histogram(field_1d, i, t_i, format):
        # Histogram
        plt.figure(figsize=(8,6))
        plt.hist(field_1d, bins = 60, histtype = "step", label = "Gas density $\\rho \\,[g/cm^3]$")
        plt.xlabel("Intensity")
        plt.ylabel('Frequency')
        # Adding a box with the time-stamp
        plt.text(0.05, 0.95, f"Time: {t_i:.3e} [s]", transform=plt.gca().transAxes,
                 fontsize=10, verticalalignment='top',
                 bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        plt.legend()
        plt.savefig(f"output_data/density_{i:03d}_histogram.{format}")
        plt.close()
        return
        
    for i in range(0, len(t_array)):
        filename = "data.0{:03d}.vtk".format(i)
        rho_i, vx1_i, vx2_i, Bx1_i, Bx2_i, prs_i, mesh_i, t_i  = io_vtk(directory,filename)
        
        # Let's plot the density
        plotting_scalar(rho_i, mesh_i, i, t_i, "density", format)
        
        # For the kinetic energy density
        # First, let's get the velocity magnitud
        v_i = np.sqrt(vx1_i**2 + vx2_i**2)
        E_i = (rho_i*(v_i**2))/2
        
        # Now, let's plot the kinetic energy density
        plotting_scalar(E_i, mesh_i, i, t_i, "kinetic_energy", format)
        
        # Let's make the histograms, first we need to flatten the density 2D array
        # We could take the log of it to have smaller values, however we want to see if these follow an specific statistical distribution.
        rho_i_1d = rho_i.flatten()
        # And now the histogram
        histogram(rho_i_1d, i, t_i, format)
    return
```

(f) Briefly describe: what happens with the density field as time progresses? Does the density field follow any statistical distribution at late times?

*Answer:*<br>
From the histograms, we could see a normal distribution at late times! 
$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2} $$
However, it also seems like before this normal distribution, we have a log normal distribution. <br>
Now, we can perform a quick regression to prove it!

(gx2) Create a set of Python functions that loops over all the VTK simulation files, computes the following quantities in CGS units for each time:

- the average gas temperature, $\overline T$, (**Hint:** the temperature in each grid cell can be calculated using the equation of state for ideal gases, i.e., $p=\frac{\rho k_B\,T}{\mu m_u}$, where $k_B$ is the Boltzmann constant, $m_u$ is the atomic mass unit, and $\mu=0.6$ is the mean particle mass in the gas.)

- the average kinetic energy density, $\overline E_k$, where $E_k = \frac{1}{2}\rho\,v^2$.

- the average magnetic energy density, $\overline E_m$, where $E_m = \frac{1}{2}\frac{B^2}{\,\mu_0}$, where $\mu_0\equiv$ magnetic permeability of free space.

and returns:

- a CSV file with 4 columns, time on the first column, and the above quantities in the next ones. The CSV file should be named "stats.csv" saved into the folder called "output_data".



*Answer:*<br>
So, for the average gas temperature $$ T =\frac{p \, \mu\, m_u}{\rho k_B}$$
And, as we're working in CGS units, we get $$E_m = \frac{B^2}{8\,\pi}$$

```python
# Defining our mean particle mass, atomic mass unit, kb constant and mu_0
mu = 0.6
m_u = cte.atomic_mass*(10**3) # in grams (cgs units)
kb = cte.Boltzmann* (10**7) # in erg/K (cgs units)
mu_0 = 1 # in cgs is dimentionless
```

```python
def quantities_to_csv(directory):
    """
    This functions opens and reads multiple vtks files and saves the average gas temperature, average kinetic energy density,
    magnetic energy density quantities for each time step in a csv file.
    Input:
        directory (str) -> location of the .vtk files
    Output:
        stats.csv (csv file) -> csv file with dataframe of all the quantities. Saved at "output_data" folder
    Author: R.S.S.G
    Date created: 19/10/2024
    """
    # Empty arrays, ready to get the values for each time
    mean_T = []
    mean_Ek = []
    mean_Em = []
    
    for i in range(0, len(t_array)):
        filename = "data.0{:03d}.vtk".format(i)
        rho_i, vx1_i, vx2_i, Bx1_i, Bx2_i, prs_i, mesh_i, t_i  = io_vtk(directory,filename)
        
        # Now, the temperature
        T_i = (prs_i*mu*m_u)/(rho_i*kb) # in K (cgs units)
        # And the mean of the grid will be
        T_mean_i = np.mean(T_i)
        # Now, append it to the tracking array
        mean_T.append(T_mean_i)
        
        # For the kinetic energy density
        v_i = np.sqrt(vx1_i**2 + vx2_i**2)
        Ek_i = (rho_i*(v_i**2))/2
        # And the mean
        Ek_mean_i = np.mean(Ek_i)
        mean_Ek.append(Ek_mean_i)
        
        # For the magnetic energy density
        B_i = np.sqrt(Bx1_i**2 + Bx2_i**2)
        Em_i = (B_i**2)/(8*np.pi)
        # And the mean
        Em_mean_i = np.mean(Em_i)
        mean_Em.append(Em_mean_i)
    
    mean_T = np.array(mean_T)
    mean_Ek = np.array(mean_Ek)
    mean_Em = np.array(mean_Em)

    df_dic = {"Time": t_array, "Average Gas Temperature": mean_T, "Average Kinetic Energy Density": mean_Ek,\
              "Average Magnetic Energy Density": mean_Em}
    df = pd.DataFrame(df_dic)
    df.to_csv("output_data/stats.csv", index = False, float_format = "{:6e}".format)
    return
```

```python
def plotting_quantities(filename, format):
    """
    This functions opens and reads a csv file containing the average gas temperature, average kinetic energy density,
    magnetic energy density and saves plots of these quantities at "output_data" folder.
    Input:
        filename (str) -> location and name of the .csv file
        format (str) -> Specific format for the images.
    Output:
        average_temperature.eps (eps image) -> Plot of mean temperature vs time.
        average_kinetic_energy_density.eps (eps image) -> Plot of mean kinetic energy density vs time.
        average_magnetic_energy_density.eps (eps image) -> Plot of mean magnetic energy density vs time.
        Additionally, for plotting in future items:
        t (np.array (of) floats) -> time array
        mean_T (np.array (of) floats) -> Average Gas Temperatur
        mean_Ek (np.array (of) floats) -> Average Kinetic Energy Density
        mean_Em (mp.array (of) floats) -> Average Magnetic Energy Density
    Author: R.S.S.G
    Date created: 19/10/2024
    """
    # Let's define an overall plotting function
    def plotting(name, quantity, format):
        """
        This function creates a plot for the specified physical quantity over time and saves it as a file.
        Input:
            name (str)-> The name of the physical quantity to plot. Options are:
                - "temperature": Average gas temperature
                - "kinetic_energy_density": Average kinetic energy density
                - "magnetic_energy_density": Average magnetic energy density
            quantity (np.array of floats): The values of the quantity to be plotted against time.
            format (str): The file format for saving the plot (e.g., "png", "pdf").
        Output:
            The function saves the plot as an image file in the specified format.
        Author: R.S.S.G
        Date created: 19/10/2024
        """
        plt.figure(figsize = (8,6))
        
        if name == "temperature":
            plt.title("Average Gas Temperature vs Time")
            plt.plot(t_array, quantity, label = "Average Gas \n Temperature $\\overline{T}$", color = "k")
            plt.ylabel("$T \\, [K]$")
        elif name == "kinetic_energy_density":
            plt.title("Average Kinetic Energy Density vs Time")
            plt.plot(t_array, quantity, label = "Average Kinetic \n Energy Density $\\overline{E_k}$", color = "orange")
            plt.ylabel("$E \\, [erg/cm^3]$")
        else:
            plt.title("Average Magnetic Energy Density vs Time")
            plt.plot(t_array, quantity, label = "Average Magnetic \n Energy Density $\\overline{E_m}$", color = "b")
            plt.ylabel("$E \\, [erg/cm^3]$")
        
        plt.xlabel("$t \\, [s]$")
        plt.legend()
        plt.savefig(f"output_data/average_{name}.{format}")
        plt.close()
    
    # Opening the csv file
    df_stats = pd.read_csv("/home/rolando/Downloads/Yachay/7th Semester/computational_physics/exams/output_data/stats.csv")
    # Extracting the quantities
    t = np.array(df_stats["Time"])
    mean_T = np.array(df_stats["Average Gas Temperature"])
    mean_Ek = np.array(df_stats["Average Kinetic Energy Density"])
    mean_Em = np.array(df_stats["Average Magnetic Energy Density"])
    # Making the plots
    plotting("temperature", mean_T, format) # For the average temperature
    plotting("kinetic_energy_density", mean_Ek, format) # For the average kinetic energy density
    plotting("magnetic_energy_density", mean_Em, format) # For the average magnetic energy density

    return t, mean_T, mean_Ek, mean_Em
```

Now, let's analyze it a little bit more by getting the thermal energy associated to the average gas temperature.<br>
So, for a monoatomic ideal gas, we know $$U = \frac{3}{2}N k_B T. $$
Then, we want the energy density $$u = \frac{3}{2}\frac{N}{V} k_B T = \frac{3}{2} \frac{\rho \, k_B \,T}{\mu \, m_u}$$ 
Using the equation of ideal gases and as we want the average thermal energy density 
$$\overline{u} = \frac{3}{2} \overline{p}$$

```python
# Let's create a function that calculates the average pressure!
# Basically the same thing that we did on previous items.
def mean_pressure(directory):
    """
    This functions opens and reads multiple vtks files and saves the average pressure
    Input:
        directory (str) -> location of the .vtk files
    Output:
        avg_p (np.array (of) floats) -> array with the average pressure in the grid for each time step.
    Author: R.S.S.G
    Date created: 19/10/2024
    """
    # Empty arrays, ready to get the values for each time
    avg_p = []
   
    for i in range(0, len(t_array)):
        filename = "data.0{:03d}.vtk".format(i)
        rho_i, vx1_i, vx2_i, Bx1_i, Bx2_i, prs_i, mesh_i, t_i  = io_vtk(directory,filename)
        # Getting the average pressure
        mean_p_i = np.mean(prs_i)
        avg_p.append(mean_p_i)
    
    avg_p = np.array(avg_p)
    return avg_p
```

```python
avg_p = mean_pressure(directory)
```

```python
# Calculating the thermal energy density
u = avg_p*(3/2)
```

(j) Create a Python function that returns movies showing the time evolution of the kinetic energy density maps computed in (d) and their average values calculated in (g). The movies should be saved into the folder called "output_data". 


```python
def movies(directory, mean_Ek):
    """
    This function creates two types of animations based on kinetic energy density maps and the average kinetic energy over time.
    Input:
        directory (str) -> location of the .png files and output directory for saving GIFs.
        mean_Ek (np.array (of) floats) -> array containing the average kinetic energy at each time step.
    Output:
        kinetic_gif_display (IPython.display.Image) -> GIF displaying the kinetic energy density maps.
        kinetic_average_display (IPython.display.HTML) -> HTML animation showing the average kinetic energy vs. time.
    Author: R.S.S.G
    Date created: 20/10/2024
    """
    def kinetic_map_animation(directory):
        """
        This function creates an animation from a series of PNG images representing kinetic energy density maps
        and saves it as a GIF.
        
        Input:
            directory (str): The directory containing the input PNG images for the animation.
        Output:
            IPython.display.Image: The GIF displaying the kinetic energy density maps.
        Author: R.S.S.G
        Date created: 20/10/2024
        """
        # Animation for the kinetic energy density maps
    
        images_input = directory + "kinetic_energy_***_map.png"        
        imgif_output = directory + "kinetic_energy.gif"
        
        # Collect the images
        imgs = (Image.open(f) for f in sorted(glob.glob(images_input)))

        # Getting the animation
        img = next(imgs)
        img.save(fp = imgif_output, format="GIF", append_images=imgs, save_all=True, duration = 100, loop = 0, optimize=True,dither=Image.NONE)
    
        return display_image(open(imgif_output, "rb").read())
    
    def kinetic_average_animation(mean_Ek):
        """
        This function creates an animation showing the average kinetic energy vs. time and saves it as a GIF.
        
        Input:
            mean_Ek (np.array (of) floats) -> array containing the average kinetic energy for each time step.
        Output:
            IPython.display.HTML -> HTML animation showing the average kinetic energy vs. time.
        Author: R.S.S.G
        Date created: 20/10/2024
        """
        
        # Create figure for the animation
        fig, ax = plt.subplots(1, 1, figsize=(10, 5))
        point, = ax.plot([], [], marker="o", color="r")
        ax.set_xlim(t_array[0], t_array[-1])
        ax.set_ylim(0., 3 * (10**(-14)))
        
        plt.title("Average Kinetic Energy Density vs Time")
        plt.plot(t_array, mean_Ek, label="Average Kinetic \n Energy Density $\\overline{E_k}$", color="purple", linestyle="--")
        plt.ylabel("$E \\, [erg/cm^3]$")
        plt.xlabel("$t \\, [s]$")
        plt.legend()
        
        # Function to animate the point
        def animate_point(i):
            x_i = t_array[i]
            y_i = mean_Ek[i]
            point.set_data([x_i], [y_i])  # Pass as sequences (list)
            return point,
        
        # Create the animation
        movie_wave = animation.FuncAnimation(fig, animate_point, frames=len(t_array), interval=100, blit=True)
        
        # Save the animation as GIF
        gif_output = directory + "kinetic_energy_animation.gif"
        movie_wave.save(gif_output, writer="pillow", fps=30)
        
        plt.close()
        
        return HTML(movie_wave.to_jshtml())

    # Run the animations and return their visualizations
    kinetic_gif_display = kinetic_map_animation(directory)
    kinetic_average_display = kinetic_average_animation(mean_Ek)
    
    return kinetic_gif_display, kinetic_average_display
```
